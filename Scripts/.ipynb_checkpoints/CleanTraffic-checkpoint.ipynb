```python
# Dependencies
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import sqlite3
```

```python
# Name of the CSV file
file = '../Resources/Police_Incidents.csv'
```

```python
#Reads in the File as a dataframe
po_poDF = pd.read_csv(file);
```

```python
# Creates a function that pulls the coordinates out of a string and returns the coordinate.
def ConvertCoor(text):
    Coordinates=text[text.find("(")+1:text.find(")")]
    Coordinates = Coordinates.split(',')
    try:
        Coordinates = [float(Coordinates[0]),float(Coordinates[1])]
    except:
        Coordinates = np.nan
    return Coordinates
```

```python
po_poDF.head()
```

```python
#Drop any rows with a NaN in the Location column.
po_poDF.dropna(subset=['Location1'],inplace=True)

#Pull out the columns of interest to another dataframe.
cleanPo_poDF=po_poDF[['Type of Incident','Type  Location','Year of Incident', 'Location1', 'Zip Code','Update Date']].copy()
cleanPo_poDF.head()
```

```python
#Creates a Coordinate column that uses the apply function to pull the coordinate data using the ConvertCoor function from above.
cleanPo_poDF['Coordinates']=cleanPo_poDF['Location1'].apply(ConvertCoor,convert_dtype=True)

```

```python
#Drops any rows where the coordinates may be missing.
cleanPo_poDF.dropna(subset=['Coordinates'],inplace=True)
```

```python

#Pulls out the Latitude and Longitude of the coordinate column.
Lat = [];
Long = [];
for coor in cleanPo_poDF['Coordinates']:
    Lat.append(coor[0])
    Long.append(coor[1])
```

```python
cleanPo_poDF['Latitude']=Lat
cleanPo_poDF['Longitude']=Long
```

```python
cleanPo_poDF.head()
```

```python
#Converts the 'Update Date' Column to a datatime data type.
cleanPo_poDF['Update Date']=pd.to_datetime(cleanPo_poDF['Update Date'])
```

```python
#Gets the day of week that the accident landed on.
cleanPo_poDF['Day of Week']=cleanPo_poDF['Update Date'].dt.day_name()
```

```python
cleanPo_poDF.head()
```

```python
#Drops the Coordinate column so that it dataframe can be ported to sqlite
cleanPo_poDF.drop(columns=['Coordinates'],inplace=True)
```

```python
cleanPo_poDF.head()
```

```python
#Outputs the final dataframe to a CSV.
cleanPo_poDF.to_csv('../Resources/CleanPo_po.csv')
```

```python
#Creates a SQLite file.
conn=sqlite3.connect('../resources/Traffic.db')
```

```python
#Ports over all dataframe info to the sqlite file.
cleanPo_poDF.to_sql('Dallas_Traffic',con=conn)
```
